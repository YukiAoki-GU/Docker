## 新しくWi-Fiを追加したい
Wi-Fiの機器は以下の手順で複数登録することができ，Wi-Fiの電波が強い順番でつながります。以下で登録方法を説明します。

- 現在，ラズパイにアクセスできている場合



````python
from IPython.display import YouTubeVideo
YouTubeVideo('oObj5LkliSA', width=400, height=225)
````

Jupyter labからターミナルを立ち上げ，

sudo raspi-config

というコマンドを打ちましょう。
その後，

1 System Options > S1 Wireless LAN > SSID > Password

でSSID(Wi-Fiの名前)とPasswordを登録してください。

- ラズパイにアクセスできない場合

````python
from IPython.display import YouTubeVideo
YouTubeVideo('Z76SJ6qa_CA', width=400, height=225)
````

ラズパイからSDカードを取り出し，SDカードにWi-Fiの情報を直接書き込みます。[こちら](https://mascii.github.io/wpa-supplicant-conf-tool/)から，

wpa_supplicant.conf

ssh.txt

の2つのファイルを作り，SDカードに書き込みます。
